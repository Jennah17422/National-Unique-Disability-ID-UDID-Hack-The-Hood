```python
#How many people have disabilities?

import pandas
import matplotlib.pyplot as plt
#How many people have disabilities?
dataFrame = pandas.read_csv("UDIDDATA.csv")

Total_people = dataFrame['total'].sum()
plt.bar(['Total People'], [Total_people], color='blue')
plt.title('Total Number of People')
plt.ylabel('Number of People')
plt.show()

Total_people
```

```python
#For my data I had to change how the age groups were presented
# 0-6 --> group 1
# 6-15 --> group 2
# 15-35 --> group 3
# 35-60 --> group 4
# 60-80 --> group 5
# 80 Year and above --> group 6


dataFrame = dataFrame.replace(['0-6 Years', '6-15 Years', '15-35 Years', '35-60 Years', '60-80 Years', '80 Year and above'], [1,2,3,4,5,6])


#How many people below the age of 15 have a disability?

under_15_df = dataFrame[dataFrame['age_group (Years)'] < 3]
total_people_under_15 = under_15_df['total'].sum()
people_under_15 = under_15_df.groupby(['age_group (Years)'])['total'].sum()


plt.bar(people_under_15.index, people_under_15, color='skyblue')
plt.title('Total Number of People Under 15')
plt.ylabel('Number of People') #In millions 
plt.xticks(range(1,3))
plt.show()


total_people_under_15
```

```python
#How many women and men have a disability?


import pandas
import matplotlib.pyplot as plt
#How many people have disabilities?
dataFrame = pandas.read_csv("UDIDDATA.csv")


Male_count = dataFrame['Male_count'].sum()
female_count = dataFrame['female_count'].sum()
plt.bar(["male", "female"], [Male_count, female_count], color=['blue', 'pink'])
plt.show()
Male_count
```

```python
#What state are the people with disabilities from?
state_counts = dataFrame['state_name'].value_counts()
plt.figure(figsize=(12,6))
state_counts.plot(kind= 'bar', color= 'green')

plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
#What are the type of disability?

disabilities = dataFrame['disability_type'].value_counts()
plt.figure(figsize=(12,6))
disabilities.plot(kind= 'bar', color= 'teal')

plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
#What are the top 5 districts for people with disabilities from?

district_counts = dataFrame['district_name'].value_counts()
top_5_districts = district_counts.head(5)
plt.figure(figsize=(12,8))
top_5_districts.plot(kind= 'bar', color= 'orange')

plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Total Patients
 
Age_groups = dataFrame.groupby('age_group (Years)')['total'].sum()
plt.figure(figsize=(16,6))
plt.pie(Age_groups, labels=Age_groups.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.show()
```
